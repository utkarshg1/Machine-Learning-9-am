# Logistic Regression

```python
from warnings import filterwarnings
filterwarnings('ignore')
```

### Step 1 : Read the dataset

```python
import pandas as pd
df = pd.read_csv('iris.csv')
df.head()
```

### Step 2: Perform basic data quality checks

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

### Step 3 : Seperate X and Y

```python
X = df.drop(columns=['species'])
Y = df[['species']]
```

```python
X.head()
```

```python
Y.head()
```

```python
Y.value_counts()
```

```python
Y.value_counts().plot(kind='bar')
```

### Step 4: Apply preprocesssing on X

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
num_pipe = Pipeline(steps=[('impute', SimpleImputer(strategy='mean')),
                           ('scaler', StandardScaler())]).set_output(transform='pandas')
```

```python
X_pre = num_pipe.fit_transform(X)
X_pre.head()
```

### Step 5: Apply Train Test split

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X_pre, Y, test_size=0.33, random_state=21)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
xtest.head()
```

```python
ytest.head()
```

```python
xtrain.shape
```

```python
xtest.shape
```

### Step 6: Model Building

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(xtrain, ytrain)
```

```python
model.classes_
```

```python
model.intercept_
```

```python
model.coef_
```

### Step 7: Model Evaluation

```python
# Predict the results for train and test
ypred_train = model.predict(xtrain)
ypred_test = model.predict(xtest)
```

```python
ypred_train[0:5]
```

```python
ytrain.head()
```

```python
ypred_test[0:5]
```

```python
ytest.head()
```

```python
model.classes_
```

```python
# Plot the confusion matrix for test data
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cf = confusion_matrix(ytest, ypred_test)
cfd = ConfusionMatrixDisplay(cf, display_labels=model.classes_)
cfd.plot()
```

```python
from sklearn.metrics import classification_report
print(classification_report(ytest, ypred_test))
```

```python
# Get the cross validated f1 score
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model, xtrain, ytrain, cv=5, scoring='f1_macro')
scores
```

```python
scores.mean()
```

### Predict the probability for train and test

```python
yprob_train = model.predict_proba(xtrain)
yprob_test = model.predict_proba(xtest)
```

# Step 8:Predict out of sample Data
