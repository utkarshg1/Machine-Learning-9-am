### Step 1: Data Ingestion (Read the csv file)

```python
import pandas as pd
df = pd.read_csv('Cars93.csv', na_values=['','NA'], keep_default_na=False)
df.head()
```

### Step 3 : Seperate X and Y

```python
X = df.drop(columns=['id', 'Weight'])
Y = df[['Weight']]
```

```python
X.head()
```

```python
Y.head()
```

### Step 4 : Create a preprocessing pipeline on X

```python
X.columns
```

```python
X.dtypes
```

```python
cat = list(X.columns[X.dtypes=='object'])
con = list(X.columns[X.dtypes!='object'])
```

```python
cat
```

```python
con
```

```python
from sklearn.pipeline import Pipeline # Used to automate preprocessing
from sklearn.impute import SimpleImputer # Used to replace missing values
from sklearn.preprocessing import StandardScaler, OneHotEncoder # Preprocessing 
from sklearn.compose import ColumnTransformer # Used to combine cat and con pipeline
```

```python
num_pipe = Pipeline(steps=[('impute', SimpleImputer(strategy='mean')),
                           ('scaler', StandardScaler())])
```

```python
cat_pipe = Pipeline(steps=[('impute', SimpleImputer(strategy='most_frequent')),
                           ('ohe', OneHotEncoder(handle_unknown='ignore', sparse_output=False))])
```

```python
pre = ColumnTransformer([('num', num_pipe, con),
                         ('cat', cat_pipe, cat)]).set_output(transform='pandas')
```

```python
pre
```

```python
X_pre = pre.fit_transform(X)
X_pre.head()
```

```python
X_pre.to_csv('Preprocessed.csv', index=False)
```

### Step 5: Apply Train test split
20% of data will be unseen to model

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X_pre, Y, test_size=0.2, random_state=21)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
xtest.head()
```

```python
ytest.head()
```

### Preprocessing is completed

### Step 6 : Built the ML model

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(xtrain, ytrain)
```

### Step 7 : Evaluate the regression model
mse, rmse, r2, mae, mape

```python
model.score(xtrain, ytrain)
```

```python
model.score(xtest, ytest)
```

```python
def evaluate_model(model, x, y):
    # Predict the results for x
    ypred = model.predict(x)
    # Import necessary packages
    from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
    from sklearn.metrics import mean_absolute_percentage_error
    # Calculate all above metrics
    mse = mean_squared_error(y, ypred)
    rmse = mse**(1/2)
    mae = mean_absolute_error(y, ypred)
    mape = mean_absolute_percentage_error(y, ypred)
    r2 = r2_score(y, ypred)
    # Print above results
    print(f'Mean Squared Error : {mse:.2f}')
    print(f'Root Mean Squared Error : {rmse:.2f}')
    print(f'Mean Absolute Error : {mae:.2f}')
    print(f'Mean Absolute Percentage Error : {mape:.4f}')
    print(f'R2 Score : {r2:.4f}')
```

```python
evaluate_model(model, xtrain, ytrain)
```

```python
evaluate_model(model, xtest, ytest)
```
